```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
mean_1 = 50
std_1 = 5
sample_size_1 = 1000
data_1 = np.random.normal(mean_1, std_1, sample_size_1)
plt.figure(figsize=(10, 6))
plt.hist(data_1, bins=30, density=True, alpha=0.6, color='g', label='Histogram of Generated Data')
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mean_1, std_1)
plt.plot(x, p, 'k', linewidth=2, label='PDF (Normal Distribution)')
plt.title('Normal Distribution (Mean=50, Std=5)')
plt.xlabel('Data Points')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python
mean_2 = 14
std_2 = 2.5
sample_size_2 = 1000
prob_1 = norm.cdf(14 + 2.5, mean_2, std_2) - norm.cdf(14, mean_2, std_2)
prob_2 = norm.cdf(12, mean_2, std_2)
prob_3 = 1 - norm.cdf(19, mean_2, std_2)
students_1 = prob_1 * sample_size_2
students_2 = prob_2 * sample_size_2
students_3 = prob_3 * sample_size_2
print(f"Students scoring between 14 and 2.5: {students_1:.2f}")
print(f"Students scoring below 12: {students_2:.2f}")
print(f"Students scoring above 19: {students_3:.2f}")
```
