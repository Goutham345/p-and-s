```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
mean = 100
standard_deviation = 15
x = np.linspace(55, 145, 500)
y = norm.pdf(x, mean, standard_deviation)
p_above_130 = 1 - norm.cdf(130, mean, standard_deviation)
p_below_85 = norm.cdf(85, mean, standard_deviation)
p_between_90_110 = norm.cdf(110, mean, standard_deviation) - norm.cdf(90, mean, standard_deviation)
plt.figure(figsize=(10, 6))
plt.plot(x, y, label="IQ Distribution", color='yellow')
x1 = np.linspace(130, 145, 200)
plt.fill_between(x1, norm.pdf(x1, mean, standard_deviation), color='skyblue', alpha=0.5, label='IQ > 130')
x2 = np.linspace(55, 85, 200)
plt.fill_between(x2, norm.pdf(x2, mean, standard_deviation), color='red', alpha=0.5, label='IQ < 85')
x3 = np.linspace(90, 110, 200)
plt.fill_between(x3, norm.pdf(x3, mean, standard_deviation), color='darkblue', alpha=0.5, label='90 < IQ < 110')
plt.title('Normal Distribution of IQ Scores')
plt.xlabel('IQ')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
print(f"Probability that IQ > 130: {p_above_130:.4f}")
print(f"Probability that IQ < 85: {p_below_85:.4f}")
print(f"Probability that 90 < IQ < 110: {p_between_90_110:.4f}")
```
